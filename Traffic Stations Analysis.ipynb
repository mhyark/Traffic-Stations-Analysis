## What was the most frequent ticket type?


```python
# Group by STOP_NAME and aggregate NB_CHECKINS
gp_stop = df_data.groupby('Station').sum()['PassengersNumber']

# DataFrame
gp_stop = pd.DataFrame(gp_stop)
gp_stop.reset_index(inplace=True)
```

```python
# DataFrame of most frequent TICKET_TYPE per STOP_NAME

# Total Passanger grouped by STOP_NAME and TICKET_TYPE
gbo_stop_tt1 = df_data.groupby(['Station', 'TicketType1'])
gp_stop_tt1 = gbo_stop_tt1.sum()['PassengersNumber']

# DataFrame of most frequent TICKET_TYPE per STOP_NAME
stop_max_tt1 = gp_stop_tt1.reset_index()
stop_max_tt1 = stop_max_tt1.sort_values('PassengersNumber', ascending=False).drop_duplicates(subset='Station').sort_index()
stop_max_tt1 = stop_max_tt1.loc[:,['Station', 'TicketType1']]
stop_max_tt1.set_index('Station', inplace=True)


# Map gp_stop (aggregated NB_CHECKINS per STOP_NAME) 
# & stop_max_tt (most frequent TICKET_TYPE per STOP_NAME)

gp_stop['MAX_TICKET_TYPE'] = gp_stop['Station'].map(stop_max_tt1['TicketType1'])
```

```python
# DataFrame of most frequent TICKET_TYPE per STOP_NAME

# Total Passanger grouped by STOP_NAME and TICKET_TYPE
gbo_stop_tt2 = df_data.groupby(['Station', 'TicketType2'])
gp_stop_tt2 = gbo_stop_tt2.sum()['PassengersNumber']

# DataFrame of most frequent TICKET_TYPE per STOP_NAME
stop_max_tt2 = gp_stop_tt2.reset_index()
stop_max_tt2 = stop_max_tt2.sort_values('PassengersNumber', ascending=False).drop_duplicates(subset='Station').sort_index()
stop_max_tt2 = stop_max_tt2.loc[:,['Station', 'TicketType2']]
stop_max_tt2.set_index('Station', inplace=True)


# Map gp_stop (aggregated NB_CHECKINS per STOP_NAME) 
# & stop_max_tt (most frequent TICKET_TYPE per STOP_NAME)

gp_stop['MAX_TICKET_TYPE'] = gp_stop['Station'].map(stop_max_tt2['TicketType2'])
```
